```python

car_price_df[['Company', 'Model']] = car_price_df['CarName'].str.extract('(\w+)\s(\w+)', expand=True)

print(car_price_df)
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=car_price_df, x='Company')
plt.xticks(rotation=90)
plt.title('Count of Cars by Company')
plt.xlabel('Company')
plt.ylabel('Count')
plt.show()
```

```python
sns.pairplot(car_price_df)
plt.show()
```

```python
numeric_cols = car_price_df.select_dtypes(include=['float64', 'int64']).columns.tolist()
numeric_cols.remove('car_ID')


sns.pairplot(car_price_df, x_vars=numeric_cols, y_vars=['price'])
plt.show()
```

```python
toyota_df = car_price_df[car_price_df['Company'] == 'toyota']

sns.pairplot(toyota_df, x_vars=numeric_cols, y_vars=['price'])
plt.show()
```

```python
normal_price_cars_df= car_price_df[car_price_df['price'] >= 20000 ]
high_price_cars_df= car_price_df[car_price_df['price'] < 20000 ]
sns.pairplot(normal_price_cars_df, x_vars=numeric_cols, y_vars=['price'])
sns.pairplot(high_price_cars_df, x_vars=numeric_cols, y_vars=['price'])

plt.show()
```

```python
numeric_df = car_price_df.select_dtypes(include=['float64', 'int64'])

corr_matrix = numeric_df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
for column in car_price_df.select_dtypes(include=['float64', 'int64']).columns:
    sns.boxplot(x=car_price_df[column])
    plt.title(f'Box plot of {column}')
    plt.show()
```
